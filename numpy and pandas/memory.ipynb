```python
import sys
cube=[i**2 for i in range(100)]
memory=sys.getsizeof(cube)*len(cube)
print(memory)
```

```python
import numpy as np
a=np.array(cube)
memory=a.itemsize*a.size
print(memory)
```

```python
import time
l=[i for i in range(10000)]
start_time=time.perf_counter()
res=l*3
end_time=time.perf_counter()
print(end_time-start_time)
```
