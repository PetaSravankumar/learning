```python
import numpy as np
l1=[[1,2,3],[4,5,6]]
l2=[[1,2],[3,4]]
l3=[[5,6],[7,8]]
a1=np.array(l1)
a2=np.array(l2)
a3=np.array(l3)
print(a1)
print(a2)
print(a3)
```

```python
res1=a1.T
print(res1)
```

```python
res2=a1.swapaxes(0,1)
print(res2)
```

```python
res3=np.dot(a2,a3)
print(res3)
```
